##**Resumen:** 
> La caída libre es cuando desde cierta altura un cuerpo se deja caer para permitir que la fuerza de gravedad actúe sobre él,la velocidad inicial es siempre cero. 
En  el  presente  informe,  se  tiene  como  fin  principal  el poder  identificar  y  comparar  los  movimientos  de “MRU” movimiento rectilíneo  uniforme  y  “MUA” movimiento uniformemente acelerado, a su vez de entender el concepto de cada uno de los dos movimientos, usando medios online como tracker para ser más exactos con las mediciones que se tomen al hacer el proceso de comparación.


##**Introducción:** 
> En caída libre, un objeto cae verticalmente desde una cierta altura, ignorando cualquier tipo de fricción con el aire o cualquier otro obstáculo, es un movimiento rectilíneo con aceleración uniforme (MRUA), donde la aceleración corresponde al valor de la gravedad, también encontramos al movimiento rectilíneo uniforme (MRUA), es aquel con velocidad constante y cuya trayectoria es una línea recta. 
Para poder entender cómo funcionan estos 3 fenómenos de la física al igual de reconocer la capacidad de relacionar y analizar datos de resultados , relacionando la teoría con 3 prácticas experimentales, en este informe se van a mostr

Para la práctica de MRUA y MRU Imágenes 1 y 2 el montaje se compone por una cámara de aire y un tubo hueco en la parte superior por donde sale el aire que es recorrido sin fricción por un “carro” debido a la acción del aire para el MRUA y el uso de un peso que hala el “carro” en el MRU.

Este registro de datos se hizo con la ayuda de un emisor de señales infrarrojas . La rejilla estaba en reposo a una altura de 20cm sostenida por un imán, y el sensor infrarrojo emitía la señal constantemente. Luego, se dejó caer la rejilla con el fin de interrumpir la señal del sensor, y en ese lapso de tiempo, se tomaban los valores.
