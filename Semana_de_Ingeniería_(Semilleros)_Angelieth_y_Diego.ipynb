##3) Expansión de células madre mesenquimales en condiciones hipóxicas en meido no xenogénico suplementado con plasma pobre en plaquetas. 

image.png

####Línea de investigación: 
 - Ingeniería en tejidos y Biomateriales 

####Grupo de investigación: 
 - Semillero medicina regenerativa UNAB

##**Resumen**: 
> Estudiantes del programa de Ingeniería Biomédica de la UNAB pertenecientes al semillero medicina regenerativa proponen un proyecto de investigación que se sustenta en el cultivo primario de células madre mesenquimales con la finalidad de disponer de dichas para desarrollar posibles aplicaciones en el campo de la terapia génica, ya que estas células resultan ser muy versátiles y de gran interés por su capacidad de autorrenovarse y diferenciarse (desarrollarse) como otro tipo de células específicas mediante el cultivo en un medio suplementado con factores de crecimiento para convertirse finalmente en células como los adipocitos, miocitos, osteoblastos, entre muchas otras células que pueden llevar a cabo la regeneración de tejidos en el cuerpo mediante su desarrollo y funciones específicas.
>
> Se plantea el desarrollo del cultivo primario de estas células mediante el cultivo in vitro con condiciones hipóxicas en un medio suplementado con plasma pobre de plaquetas, ya que es el medio de cultivo que permite la expansión de las células mesenquimales. El desarrollo de este proyecto permitirá establecer el tiempo de duplicación poblacional y el ciclo celular de las células madres mesenquimales, las cuales se tratarán de mantener sin su etapa de diferenciación para tratar de influir o “programar” su diferenciación según la suplementación con diferentes factores de crecimiento originando la célula específica que se requiera en la terapia génica.
>
> De esta manera, la realización de este proyecto reconoce que las células madre mesenquimales han tomado gran importancia en el desarrollo de métodos para la regeneración de tejidos permitiendo dar paso a nuevas alternativas para la medicina regenerativa.

##5) **Sistema de sincronía de equipos de monitoreo de variables fisiológicas.**



####Línea de investigación: 
 - Neurociencias y economía y equipos biomedicos 

####Grupo de investigación: 
 - SEBIUNAB
 
##**Resumen:** 
> Integrantes del semillero de equipos biomédicos proponen un proyecto que consiste en llevar a cabo la sincronización de equipos biomédicos con la finalidad de monitorizar funciones biológicas. Se plantea este proyecto porque se identifica que los diferentes equipos biomédicos arrojan resultados de manera individual pero se plantea una hipótesis de que la sincronización de equipos biomédicos facilitaría la toma de decisiones referente al momento de diagnosticar pacientes al relacionar resultados de dos equipos como el encefalograma junto con cardiograma para determinar si algún problema cardiaco está vinculado a fallo de funciones nerviosas. 
>
>El principal obstáculo que se encuentra para el desarrollo de este proyecto es el entorno y lenguaje de programación, que es la causa por la que no existe sincronización de equipos biomédicos ya que resulta un proceso complejo y por falta de licencias todavía más. Inicialmente se había considerado utilizar MATLAB pero por lo anteriormente mencionado se determinó que este proyecto tiene una mejor viabilidad para desarrollarse en el lenguaje de programación de C++.
