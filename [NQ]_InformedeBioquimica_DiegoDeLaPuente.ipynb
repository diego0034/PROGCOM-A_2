##**Marco teórico:**
> Los carbohidratos (hidratos de carbono) son polihidroxialdehídos, polihidroxicetonas o compuestos que producen polihidroxi aldehídos o cetonas por hidrólisis. El arroz, las patatas, el pan, el maíz, los dulces y las frutas son ricos en carbohidratos. Un hidrato de carbono puede clasificarse como un monosacárido (glucosa o fructosa); un disacárido (sacarosa o lactosa), que consiste en dos monosacáridos; o un polisacárido (almidón o celulosa), que consiste en miles de unidades de monosacáridos enlazadas entre sí. Los monosacáridos existen principalmente como estructuras cíclicas que contienen grupos hemiacetal. Estas estructuras en soluciones están en equilibrio con las correspondientes estructuras de cadena abierta. La glucosa, el principal carbohidrato en sangre y que se utiliza en las principales rutas metabólicas del organismo. [1] 
Los carbohidratos son compuestos que contienen carbono, hidrógeno y oxígeno en las proporciones 6:12:6. Durante el metabolismo se queman para producir energía, y liberan dióxido de carbono (CO2) y agua (H2O). ##Discusión de resultados:

####En la prueba de Molisch
> El que interviene principalmente en la identificación de carbohidratos es el ácido sulfúrico debido a que este hidroliza (interacción entre molécula de agua y una macromolécula) polisacáridos y disacáridos hasta convertirlos en derivados del furfural los cuales al interactuar con el naftol forman el color púrpura. 

####En la prueba de Benedict
> De las 9 muestras realizadas vemos que en 6 de ellas al reaccionar con el reactivo de Benedict aparece un precipitado naranja  indicando cuales son positivos a carbohidratos, este reactivo lo que hace es identificar azucares reductores en los carbohidratos y al aplicar calor da como resultado ese color y mas especifico.
Encontramos que este reactivo tiene entre sus componentes el citrato de sodio el cual  interactúa con el grupo aldehído (CHO) en forma de Cu y ocurre una estabilización o reducción de Cu+2 a Cu+ y esta interacción es la que ocasiona ese color ladrillo representando al óxido cuproso.

####En la prueba de Lugol
> Vemos que en ambas muestras donde se aplicó el lugol apareció un color negro oscuro indicando positivo para almidón y para la explicación de esto se tiene en cuenta el principio molecular de la reacción del yodo sabiendo que el lugol está constituido básicamente por 
agua destilada y yodo donde este último es el que ayuda al reactivo a identificar el almidón.
Cuando el yodo identifica el almidón, la amilosa que es una parte condensada en forma de cadena lineal del almidon forma helices donde las moleculas de yodo se juntan y forman cadenas de poliyoduro, formando ese color negro o azul oscuro.


**Ácido Fosfórico**
> En las 3 muestras realizadas vemos que 2 dieron positivo (Sacarosa y Fructosa) a excepción de la lactosa, o que hace este reactivo es identificar la fructosa presente en un carbohidrato y formar esa coloraciones dependiendo de la concentración de la muestra y esta es la razón por la que aunque la lactosa si es un carbohidrato no da positivo al no tener fructosa presente en su estructura.
