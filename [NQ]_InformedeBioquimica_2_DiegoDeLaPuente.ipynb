##**Introducción:** 
> El pH es algo que todos hemos escuchado, pero la principal razón del por que este es un concepto ampliamente usado radica en que su importancia consiste en que es una medida la cual es vital mantenerla controlada en muchas reacciones químicas, un gran ejemplo es todas las reacciones que ocurren en los organismos de los seres vivos las cuales trabajan a un pH en específico dependiendo de la especie. El pH también garantiza la estabilidad de productos que usamos en nuestro día a día como medicamentos, alimentos, productos de aseo, etc.
En el siguiente informe se muestran los resultados de una práctica en la que realizamos la titulación de 2 soluciones con ayuda de la medición del pH.



##**Marco teórico:**
> Cuando medimos el pH de una disolución podemos determinar su grado de alcalinidad o acidez, al igual que su concentración de hidrogeniones. Una hidrogenión es un ion positivo de Hidrógeno H+ (protones).
La fórmula matemática para calcular el pH es el logaritmo negativo en base 10 de la actividad de los iones de hidrógeno. pH = –log[H+].
El pH se mide en una escala de 0 a 14. En esta escala, un valor pH de 7 es neutro, lo que significa que la sustancia o solución no es ácida ni alcalina. Un valor pH de menos de 7 significa que es más ácida, y un valor pH de más de 7 significa que es más alcalina. En el campo de la medicina, tener un pH apropiado en la sangre y otros líquidos del cuerpo es importante para el buen funcionamiento del cuerpo.[1]
Tenemos que una titulación ácido-base consiste básicamente en determinar la concentración aproximada de una solución ácida o de una solución básica mediante la medición de pH realizando mediciones de de dicho pH.[2]
El químico Linus Pauling, ganador de dos premios Nobel, afirmaba que “mantener el cuerpo con un pH alcalino es clave para disfrutar de una buena salud.”[3]
